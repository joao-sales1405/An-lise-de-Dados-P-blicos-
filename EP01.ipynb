```python
print("Análise da quantidade de funcionários na área da saúde")
```

```python
import os
import pandas as pd
cities={}
files=os.listdir('database/csv')
for f in files:
    x=f.split('.')
    if x[1]=='csv':
        cities[f.replace('.csv','')]=[f]
for c in cities:
    df=pd.read_csv('database/csv/'+cities[c][0])
    cities[c].append(df)
```

```python
cities['sjc'][1].plot(figsize=(15,5),x ='Categoria', y='Total', kind = 'bar')
```
